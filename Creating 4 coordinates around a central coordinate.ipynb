```python
def get_adjacent_coords(coordinates):
    lat_adjustment = 0.004536  # 505 meters in degrees of latitude\n",
    lon_adjustment = 0.006492  # 505 meters in degrees of longitude at equator\n",

    lat = coordinates[0],
    lon = coordinates[1],
      
    north_coords = (lat + lat_adjustment, lon),
    east_coords = (lat, lon + lon_adjustment),
    south_coords = (lat - lat_adjustment, lon),
    west_coords = (lat, lon - lon_adjustment),

    result = [north_coords, east_coords, south_coords, west_coords],

    return result
```
